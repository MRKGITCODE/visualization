```python
#Import required libraries
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
import numpy as np
```

```python
#Set randomization seed and create two NumPy arrays containing random numbers
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
# Create a line plot and legend on the lower right
fig, ax=plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = "lower right",
           frameon = True,
           fontsize = 12,
           ncol = 2,
           shadow = True)
```

```python
#Switch legend to "upper right"
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1,label = "Person 1")
ax.plot(x,y2,label = "Person 2")
ax.legend(loc = 'upper right',
          bbox_to_anchor = (1,0))
```

```python
# Scatterplot with text inside graph
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label = "Person 1")
ax.scatter(x,y2,label = "Person 2")
ax.legend(loc = 'lower right')
ax.text(10,95,"This value is important!")
```

```python
# Formatted label
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label = "Person 1")
ax.scatter(x,y2,label = "Person 2")
ax.legend(loc = 'lower right')
ax.text(10,95,"This value is important!",
        ha = 'center',
        color = 'red',
        size = 20)
```

```python
# Illustrating grid points
fig, ax = plt.subplots ()
ax.axis([0, 10, 0, 10])

ax.text(1, 5, ". Data: (1,5)", transform = ax.transData)
ax.text(0.5, 0.1, ". AXES OBJECT: (0.5,0.1)", transform = ax.transAxes)
ax.text(0.2,0.2, ". FIGURE: (0.2, 0.2)", transform = fig.transFigure)
```

```python
# Adding a wedge style arrow
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")

ax.annotate('This is important!',
            xy=(10, 95),
            xytext=(20, 94), 
            arrowprops=dict(arrowstyle="wedge",
                            color="hotpink"))

```

```python
# Modifying the x-axis
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Modifying the x-axis to show 3 ticks
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")

ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# Rotating ticks on x-axis
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")

plt.xticks(rotation=45, ha='right')

```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right",
           frameon = True,
           fontsize = 12,
                      shadow = True)

plt.xticks(rotation=45, ha='right')
```

```python
# Labelling the x-axis
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label="Person 1")
ax.scatter(x, y2, label="Person 2")
ax.legend(loc="lower right")
 

plt.xticks(rotation=45, ha='right')
plt.xlabel('New Shiny X-Axis!', fontsize=18, fontfamily = 'serif')
```

```python
# Plotting with a styles
plt.style.use('fivethirtyeight')

fig,ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1)
ax.plot(x,y2)
```
