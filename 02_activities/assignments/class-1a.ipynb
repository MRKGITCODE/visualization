```python
#Import required libraries
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#Set randomization seed and create two NumPy arrays containing random numbers
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)
```

```python
#Create a scatterplot that's 5" wide and 3" long
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
#Convert the scatterplot into a bar graph.
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
#Convert the histogram into a line graph
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
#Convert the line graph into a histogram. 
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
#Add labels
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()

#Change font size and type

font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}


fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()


```

```python
#Shift the title to the left
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
#Convert the graph into a scatterplot with stars
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,
           y,
           marker='*',
           color = "indigo")
fig.show()
```

```python
#Add dashed line to the previous plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*', color = 'indigo', linestyle = '--', linewidth = 2)
fig.show()
```

```python
#Change color and use hex
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*', color = '#7425b9', linestyle = '--', linewidth = 2)
fig.show()
```

```python
#Added gridlines and diamond markers
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='D', 
        markersize = 5,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#fa9359' )
ax.grid(axis = 'y', color = "blue", linewidth = 2, linestyle = '-.')
fig.show()
```
